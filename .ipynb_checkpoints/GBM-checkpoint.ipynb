```python
import pandas as pd,numpy as np,matplotlib.pyplot as plt,scipy.stats as st
```

```python
df_fpt = pd.read_csv('Dữ liệu Lịch sử FPT.csv')
```

```python
def Clear_data(df,name):
    data = df.copy()
    data['Ngày'] =  pd.to_datetime(data['Ngày'], format='%d/%m/%Y')
    data[name] = data['Lần cuối'].apply(lambda x: x.replace(',','')).astype('float')
    data.drop(columns=['Lần cuối','Mở','Cao','Thấp','KL'],inplace=True)
    data.rename(columns = {'Ngày':'Day','% Thay đổi':'Change'},inplace=True)
    data['Change'] = data['Change'].apply(lambda x: x.replace('%','')).astype('float')/100
    data = data.set_index('Day')
    return data
```

```python
df_fpt = Clear_data(df_fpt,'FPT')
```

```python
log_returns = np.log(1 + df_fpt.Change)
```

```python
fig,ax = plt.subplots(2,figsize=(20,10))
plt.rcParams['font.family']='Arial'
plt.rcParams['font.size']=13
ax[0].set(title='Stock Price (8/4/2022-8/5/2023)')
ax[0].plot(df_fpt['FPT'],label='FPT')
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)
ax[0].tick_params(colors='gray')
ax[0].set_ylabel('VND',color='gray',loc='top')
ax[0].legend(frameon=False,loc='lower left',labelcolor='black')
ax[1].set(title='FPT Change (8/4/2022-8/5/2023)')
ax[1].plot(log_returns,label='Change',color='orange')
ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].tick_params(colors='gray')
ax[1].set_ylabel('%',color='gray',loc='top')
ax[1].legend(frameon=False,loc='upper left',labelcolor='black');
```

```python
t_intervals = 21
iterations = 1000
drift = np.mean(log_returns)-(np.var(log_returns)*0.5)
std = np.std(log_returns)
daily_returns = np.exp(drift + std*st.norm.ppf(np.random.rand(t_intervals,iterations)))
price_list = np.zeros_like(daily_returns)
price_list[0] = df_fpt['FPT'].iloc[0]
for t in range(1,t_intervals):
    price_list[t] = price_list[t-1]*daily_returns[t]
```

```python
plt.figure(figsize=(10,6))
plt.plot(price_list);
```

```python
prices = price_list[-1,:]
```

Trong tất cả 1000 dự đoán giá FPT sau 21 ngày nữa thì trung vị bằng 77146 (VND). Vậy dự trên 1000 dự đoán thì sau 21 phiên nữa khả năng cao giá chứng khoán FPT sẽ giảm.
